```python
#importing the required libraries

import numpy as np
import pandas as pd
```

```python
#reading the .csv files

movie=pd.read_csv('tmdb_5000_movies.csv')
```

```python
#merging the 2 .csv files

final=pd.merge(movie,credits,left_on='id',right_on='movie_id')[['id','original_title','genres','keywords','overview','cast','crew']]
```

```python
final.head()
```

```python
# checking if there are any null values
final.isnull().sum()
```

```python
#dropping the null values

final.dropna(subset=['overview'],inplace=True)
```

```python
final.isnull().sum()
```

```python
final['genres'][0]
```

```python
# converting above genres data into suitable format
import ast
genres=[]
def convert(s):
    l1=[]
    for i in ast.literal_eval(s):
        l1.append(i['name'])
        if i['name'] not in genres:
            genres.append(i['name'])
    text=l1[:]
   
    l1=[]
    return text
```

```python
final['genres']=final['genres'].apply(convert)
```

```python
final['genres'].head()
```

```python
print("Genres present in the dataset:\n",genres)
```

```python
print("Number of genres in the dataset:\n",len(genres))
```

```python
# converting keywords column into suitable format

final['keywords']=final['keywords'].apply(convert)
```

```python
final['keywords'].head()
```

```python
final['overview']=final['overview'].apply(lambda x : x.split(' '))
```

```python
final.head()
```

```python
# converting the cast column into suitable format
def fetch_cast(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
            if counter<3:
                L.append(i['name'])
            else:
                break
            counter+=1
    L1=L[:]
    L.clear()
    return L1
```

```python
final['cast']=final['cast'].apply(fetch_cast)
```

```python
final.head()
```

```python
#fecthing the movie director from the crew column
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
            break
    L1=L[:]
    L.clear()
    return L1
```

```python
final['crew']=final['crew'].apply(fetch_director)
```

```python
final.head()
```

```python
#combining the movie metadata into a final 'tags' column

final['tags']=final['genres']+final['keywords']+final['overview']+final['cast']+final['crew']
```

```python
def remove_space(L):
    L1=[]
    for i in L:
        L1.append(i.replace(' ',''))
    
    L2=L1[:] 
    L1.clear()
    return L2
```

```python
final['tags']=final['tags'].apply(remove_space)
```

```python
final.head()
```

```python
final['tags']=final['tags'].apply(lambda x:" ".join(x))
```

```python
final.head()
```

```python
final['tags']=final['tags'].apply(lambda x:x.lower())
```

```python
final.head()
```

```python
#creating a new column having title in the lower format
def convert(name):
    return name.lower()

final['lower_title']=final['original_title'].apply(convert)
```

```python
# using count vectorizer to convert text data into vector form

from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(stop_words='english',max_features=10000)
```

```python
# transforming tags column into vector form

X = cv.fit_transform(final['tags']).toarray()
```

```python
X.shape
```

```python
# using cosine similarity function to find similarity matrix

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(X)
```

# Evaluating accuracy of recommendations

```python
entered_mov="titanic"
print("Entered movie:\n",entered_mov)
recs=recommend(entered_mov)
print("\nRecommendations:\n",recs)

l1=[]

# basic recommendation function to test accuracy for a particular movie

def find_simil(x,entered_mov):
    c=0
    entered_mov=entered_mov.lower()
    for i in x:
        i=i.lower()
        if i==entered_mov:
            c+=1
    print(c)
    l1.append(c)
    
for movie in recs:
        print("\nCurrent movie:\n",movie)
        recc=recommend(movie)
        print("\nRecommendations for current movie:\n",recc)
        find_simil(recc,entered_mov)
        

 
print("\nAccuracy:\n",(sum(l1)/5)*100,"%")
```

```python
# testing accuracy of recommendations for various movie inputs 

input_movies=['titanic','iron man','batman',"baby's day out",'jurassic park','x-men','the amazing spider-man','frozen','up','captain america: civil war','avengers:age of ultron','Harry potter and the goblet of fire']

# function to find overall accuracy based on various movie inputs

def find_simil(x,entered_mov):
   
    c=0
    entered_mov=entered_mov.lower()
    for i in x:
        i=i.lower()
        if i==entered_mov:
            c+=1
    print("\nMatch found with entered movie:",c)
    l1.append(c)
    return l1

accuracy_list=[]
for i in range(0,len(input_movies)):
    l1=[]
    print("\nEntered movie:\n",input_movies[i])
    recs=recommend(str(input_movies[i]))
    print("\nRecommendations:\n",recs)
    for j in range(len(recs)):
            print("\nCurrent movie:\n",recs[j])
            recc=recommend(recs[j])
            print("\nRecommendations for current movie:\n",recc)
            l1=find_simil(recc,input_movies[i])
    acc=(sum(l1)/5)
    accuracy_list.append(acc)
        

overall_acc=round((sum(accuracy_list)/12)*100,2)
print("\nOverall accuracy:\n",overall_acc,"%")
```
